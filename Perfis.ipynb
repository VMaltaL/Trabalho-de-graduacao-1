```python
points_1, bathymetry_1 = pd.read_csv("perfil_1.csv").to_numpy().T
points_2, bathymetry_2 = pd.read_csv("perfil_2.csv").to_numpy().T
points_3, bathymetry_3 = pd.read_csv("perfil_3.csv").to_numpy().T
points_4, bathymetry_4 = pd.read_csv("perfil_4.csv").to_numpy().T
points_5, bathymetry_5 = pd.read_csv("perfil_5.csv").to_numpy().T
points_6, bathymetry_6 = pd.read_csv("perfil_6.csv").to_numpy().T
```

```python
fig,figure = plt.subplots(3,2, sharex=True, sharey=True, figsize=(24,9))
plt.subplots_adjust(hspace = 0.5)

figure[0,0].plot(points_1, bathymetry_1, 'C0', label="Perfil 1", linewidth=2.25)
figure[1,0].plot(points_2, bathymetry_2, 'C1', label="Perfil 2", linewidth=2.25)
figure[2,0].plot(points_3, bathymetry_3, 'C2', label="Perfil 3", linewidth=2.25)
figure[0,1].plot(points_4, bathymetry_4, 'C3', label="Perfil 4", linewidth=2.25)
figure[1,1].plot(points_5, bathymetry_5, 'C4', label="Perfil 5", linewidth=2.25)
figure[2,1].plot(points_6, bathymetry_6, 'C5', label="Perfil 6", linewidth=2.25)

for ax in figure.flat:
  # Grade clara
  ax.grid(linestyle="--", color="lightgray")

  # Legenda
  ax.legend(fontsize=17.5)

  # Molduras (spines) finas e claras
  for spine in ax.spines.values():
    spine.set_linewidth(0.8)

  # Ticks (marcas dos eixos) escuros
  ax.tick_params(axis='both', colors='black', )  # eixo x e y em preto

  # TÃ­tulo dos eixos escuros (caso tenha)
  ax.xaxis.label.set_color('black')
  ax.yaxis.label.set_color('black')

nome ='apresentacao_perfis_sep'
fig.savefig(f"{nome}.pdf")
fig.savefig(f"{nome}.png")
```
